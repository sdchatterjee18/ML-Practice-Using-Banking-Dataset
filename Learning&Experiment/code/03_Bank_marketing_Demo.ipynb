```python
import pandas as pd
df=pd.read_csv('bank-additional-full.csv',sep=';')
print(df['y'].unique())
```

```python
for col in df.select_dtypes(include='object').columns:
    df[col]=df[col].replace('unknown',df[col].mode()[0])
print(df['y'].unique())
```

```python
df['y']=df['y'].replace('yes',1)
df['y']=df['y'].replace('no',0)
print(df['y'].unique())
```

```python
df=df.drop('duration',axis=1)
```

```python
print(df.columns)
```

```python
print(df['y'])
```

```python
print(df['job'].unique())
```

```python
df=pd.get_dummies(df,drop_first=True)
```

```python
print(df.columns)
```

```python
print(df.select_dtypes(include='boolean').columns)
```

```python
x=df.drop('y',axis=1)
y=df['y']
```

```python
print(x.columns)
print(y.head())
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42,stratify=y)
```

```python
print(x_train)
print(x_test)
print(y_train)
print(y_test)
```

```python
from sklearn.preprocessing import StandardScaler
scalar=StandardScaler()
x_train=scalar.fit_transform(x_train)
x_test=scalar.transform(x_test)
```

```python
from sklearn.tree  import DecisionTreeClassifier
model=DecisionTreeClassifier()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score,roc_auc_score
cm=confusion_matrix(y_test,y_pred)
print(cm)
accurancy=accuracy_score(y_test,y_pred)
precision=precision_score(y_test,y_pred)
recall=recall_score(y_test,y_pred)
f1=f1_score(y_test,y_pred)
roc_auc=roc_auc_score(y_test,y_pred)
print("accurancy=",accurancy)
print("precision=",precision)
print("recall=",recall)
print("f1=",f1)
print("roc_auc=",roc_auc)
```

```python
from sklearn.ensemble import RandomForestClassifier
model_rf=RandomForestClassifier()
model_rf.fit(x_train,y_train)
y_pred_rf=model_rf.predict(x_test)
```

```python
cm_rf=confusion_matrix(y_test,y_pred_rf)
print(cm_rf)
accurancy_rf=accuracy_score(y_test,y_pred_rf)
precision_rf=precision_score(y_test,y_pred_rf)
recall_rf=recall_score(y_test,y_pred_rf)
f1_rf=f1_score(y_test,y_pred_rf)
roc_auc_rf=roc_auc_score(y_test,y_pred_rf)
print("accurancy=",accurancy_rf)
print("precision=",precision_rf)
print("recall=",recall_rf)
print("f1=",f1_rf)
print("roc_auc=",roc_auc_rf)
```

```python
from sklearn.linear_model import LogisticRegression

lr_model = LogisticRegression()

lr_model.fit(x_train, y_train)

y_pred_lr = lr_model.predict(x_test)
```

```python
cm_lr=confusion_matrix(y_test,y_pred_lr)
print(cm_lr)
accurancy_lr=accuracy_score(y_test,y_pred_lr)
precision_lr=precision_score(y_test,y_pred_lr)
recall_lr=recall_score(y_test,y_pred_lr)
f1_lr=f1_score(y_test,y_pred_lr)
roc_auc_lr=roc_auc_score(y_test,y_pred_lr)
print("accurancy=",accurancy_lr)
print("precision=",precision_lr)
print("recall=",recall_lr)
print("f1=",f1_lr)
print("roc_auc=",roc_auc_lr)
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)

X_train_sm, y_train_sm = smote.fit_resample(x_train,y_train)
```

```python
from sklearn.linear_model import LogisticRegression

lr_model = LogisticRegression()

lr_model.fit(X_train_sm,y_train_sm)

y_pred_lr = lr_model.predict(x_test)
```

```python
cm_lr=confusion_matrix(y_test,y_pred_lr)
print(cm_lr)
accurancy_lr_sm=accuracy_score(y_test,y_pred_lr)
precision_lr_sm=precision_score(y_test,y_pred_lr)
recall_lr_sm=recall_score(y_test,y_pred_lr)
f1_lr_sm=f1_score(y_test,y_pred_lr)
roc_auc_lr_sm=roc_auc_score(y_test,y_pred_lr)
print("accurancy=",accurancy_lr_sm)
print("precision=",precision_lr_sm)
print("recall=",recall_lr_sm)
print("f1=",f1_lr_sm)
print("roc_auc=",roc_auc_lr_sm)
```
